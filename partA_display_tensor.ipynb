DISPLAYING THE PREPROCESSED PART_A TENSORS

```python
import torch
from pathlib import Path
import numpy as np
```

```python
IMG_DIR = Path(r"C:\Users\mahal\OneDrive\Desktop\DL\torch_images_trainA")
GT_DIR  = Path(r"C:\Users\mahal\OneDrive\Desktop\DL\torch_density_trainA")
```

```python
sample = "IMG_10"
print("Selected Sample:", sample)
```

```python
# Load tensors
img_t = torch.load(IMG_DIR / f"{sample}.pt")   # [3, H, W]
gt_t  = torch.load(GT_DIR  / f"{sample}.pt")   # [1, h, w]
```

```python
# Convert to numpy
img_np = img_t.numpy()
gt_np  = gt_t.numpy()
```

---------------------- DISPLAY HEADS ----------------------

```python
print("\n=== IMAGE TENSOR SHAPE ===")
print(img_np.shape)
```

```python
print("\n=== IMAGE TENSOR 2x2 HEAD (for each channel) ===")
print(img_np[:, :5, :5])        # (3, 2, 2)
```

```python
print("\n=== GT TENSOR SHAPE ===")
print(gt_np.shape)
```

```python
print("\n=== GT TENSOR 2x2 HEAD ===")
print(gt_np[:, :5, :5])         # (1, 2, 2)
```
