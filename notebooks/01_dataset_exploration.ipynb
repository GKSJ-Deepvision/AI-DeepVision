```python
import os
from glob import glob
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
print("Notebook ready")
```

```python
data_dir = "../data/ShanghaiTech/part_A_train"
img_paths = sorted(glob(os.path.join(data_dir, "*.jpg")))
len(img_paths)
```

```python
for p in img_paths[:4]:
    img = cv2.imread(p)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6,4))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```
