```python
import cv2
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

img_path = r"C:/Users/Lenovo/Downloads/deep dataset/part_A_final/test_data/images/IMG_20.jpg"
gt_path  = r"C:/Users/Lenovo/Downloads/deep dataset/part_A_final/test_data/ground_truth/GT_IMG_20.mat"

img = cv2.imread(img_path)
if img is None:
    raise Exception("Image not found")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

data = scipy.io.loadmat(gt_path)
points = data["image_info"][0][0][0][0][0]

density = np.zeros((img.shape[0], img.shape[1]), dtype=np.float32)
for p in points:
    x, y = int(p[0]), int(p[1])
    if 0 <= y < density.shape[0] and 0 <= x < density.shape[1]:
        density[y, x] = 1.0

density = gaussian_filter(density, sigma=15)

plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plt.imshow(img)
plt.axis("off")
plt.title("Image")

plt.subplot(2,1,2)
plt.imshow(density, cmap="jet")
plt.axis("off")
plt.title("Density Map")
plt.colorbar()

plt.show()

print("People Count =", int(density.sum()))
```
