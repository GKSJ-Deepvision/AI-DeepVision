```python
import os

# Path to your dataset folder
dataset_path = r"C:\Users\nshre\OneDrive\Desktop\DeepVision\dataset\crowd"

# Check if path exists
if not os.path.exists(dataset_path):
    print("❌ ERROR: Dataset path not found!")
else:
    print("✔ Dataset path exists.")

    # List all files
    all_files = os.listdir(dataset_path)
    print("Total files found in folder:", len(all_files))

    # Filter only image files
    valid_extensions = [".jpg", ".jpeg", ".png"]

    images = [
        f for f in all_files
        if f.lower().endswith(tuple(valid_extensions))
    ]

    print("Total valid images found:", len(images))

    # Show first 5 filenames
    print("\nSample image names:")
    print(images[:5])
```
